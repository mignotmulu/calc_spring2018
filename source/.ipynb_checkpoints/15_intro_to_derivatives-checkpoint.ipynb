# The Derivative

A **DEFINITION**:

$$f'(x) = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h}$$

```python
def f(x):
    return x**2
```

```python
x = np.linspace(-2,2,100)
```

```python
def df(f, a):
    return (f(a + h) - f(a))/h
```

```python
h = 0.001
```

```python
f(2), df(f, 3)
```

```python
plt.figure()
plt.plot(x, f(x))
plt.plot(x, df(f, x))
plt.axhline(color = 'black')
plt.axvline(color = 'black')
```

```python
def derv_plots(x, f):
    plt.figure()
    plt.plot(x, f(x), label = '$f(x)$')
    plt.plot(x, df(f, x), label = "$f'(x)$")
    plt.axhline(color = 'black')
    plt.axvline(color = 'black')
    plt.legend(loc = 'best', frameon = False)
```

```python
def g(x):
    return np.sin(x)

x2 = np.linspace(-2*np.pi, 2*np.pi, 1000)
```

```python
derv_plots(x2, g)
```

```python
def p(x):
    return x**3 - np.sin(x**2)
```

```python
derv_plots(x, p)
```

```python
def sec_line(a):
    return f(a) + (f(a) - f(b))/(a - b)*(x - a)
```

```python
plt.figure()
plt.plot(x, f(x))
plt.plot(x, sec_line(0.5))
```

```python
def derv(a, h):
    plt.figure()
    plt.plot(a, f(a), 'o', markersize = 10)
    plt.plot((a+h), f(a+h), 'o', markersize = 10)
    slope = (f(a+h) - f(a))/(h)
    def tan(x):
        return slope*(x - a) + f(a)
    plt.plot(x, tan(x))
    plt.plot(x, f(x))
    plt.title("This is a demonstration of the Derivative", loc = 'left', fontsize = 17)
```

```python
interact(derv, a = (-2,2,0.1), h = (-2, 2, 0.1));
```

Consider $f(x) = x^n$ for any integer $n$.  What is the derivative of $f$, and why?

```python
from itertools import combinations
```

```python
combos = combinations([1,2,3],2)
```

```python
for i in combos:
    print(i)
```

```python
combos = []
for i in range(100):
    lists = combinations([1, 2, 3, 4], i)
    combos.append(lists)
```

```python
def pascal(n):
    combos = []
    for i in range(n):
        lists = combinations([i for i in range(1, n)], i)
        combos.append(lists)
        for i in combos:
            for j in i:
                a = []
                a.append(j)
                print(j)
```

```python
interact(pascal, n = widgets.IntSlider(min = 0 , max = 10, step = 1, value = 3));
```

```python
x, y = sy.symbols('x y')
```

```python
def expression_expand(n):
    for i in range(n):
        sy.pprint(sy.expand((x+y)**i))
```

```python
interact(expression_expand, n = widgets.IntSlider(min = 0, max = 10, step = 1))
```
