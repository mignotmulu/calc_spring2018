# Parametric Equations and Derivatives

Suppose we have a particle whose motion is described by the following parametric equations.

$$x(t) = 2t - 4$$
$$y(t) = 3 + t^2$$

We might want to determine the following:

- Describe the particles motion on some interval
- Determine information about the velocity and acceleration of the object

We can use familiar ideas about derivatives here, and recognize that the parametric quotient is what we're looking for here.

$$\frac{dy}{dx} = \frac{dy/dx}{dx/dt} = \frac{y'(t)}{x'(t)}$$

**PROBLEMS**

1. Describe the position of the particle at $t = 1, 2, 3, 4, 5$
2. What is the velocity of the particle at time $t = 5$

```python
t = np.arange(0, 20)
x = 2*t - 4
y = 3 + t**2
```

```python
plt.plot(x, y, '-->')
```

### Ellipses Etc.

We can also examine conic sections represented parametrically.  Here, we can represent an ellipse as

$$x(t) = x_c + r\cos(\theta)$$
$$y(t) = y_c + r\sin(\theta)$$

where $(x_c, y_c) $ is the center and $r$ the radius in given direction.

```python
theta = np.linspace(0, 2*np.pi, 30)
x = 2 + np.cos(theta)
y = 2 + np.sin(theta)
```

```python
plt.plot(x, y, '-->')
```
