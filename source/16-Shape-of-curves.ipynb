```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Derivatives and Shapes of Curves

```python
# defines a derivative
def derv(x):
    return (f(x+0.000001) - f(x))/0.000001
```

```python
def f(x):
    return x**2
```

```python
x = np.linspace(-2, 2, 100)
```

```python
# plot of function and its derivative
plt.plot(x, f(x), label = '$f(x)$')
plt.plot(x, derv(x), label = "$f'(x)$")
plt.legend(loc = 'best')
plt.axhline(color = 'black')
plt.axvline(color = 'black')
```

### Making a Function

Suppose we wanted to make a function with a few zeros between $x = [-3, 3]$. 

```python
def f(x):
    return x*(x - 2)*(x - 1)*(x + 1)*( x + 2)
```

```python
plt.plot(x, f(x), label = '$f(x)$')
plt.plot(x, derv(x), label = "$f'(x)$")
plt.legend(loc = 'best')
plt.axhline(color = 'black')
plt.axvline(color = 'black')
```

### A Function for Plotting Both

```python
def pair_plot(x, f):
    plt.plot(x, f(x), label = '$f(x)$')
    plt.plot(x, derv(x), label = "$f'(x)$")
    plt.legend(loc = 'best')
    plt.axhline(color = 'black')
    plt.axvline(color = 'black')
    plt.ylim(-5,5)
```

```python
x = np.linspace(-4,4,100)
def f(x): return x*(x - 1)*(x + 1)
pair_plot(x, f)
```

```python
x = np.linspace(-4,4,100)
def f(x): return x*np.sin(x)
pair_plot(x, f)
```
